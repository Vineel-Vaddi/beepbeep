```python
from pymongo import MongoClient
from pymongo.errors import PyMongoError

# Paste your MongoDB connection string here temporarily for local testing
MONGO_URI = "mongodb+srv://techwithaibuddies_db_user:FcbzZS0tRfMDEBDO@cluster0.naa9gp.mongodb.net/"


def main():
    try:
        client = MongoClient(MONGO_URI)

        # Quick connection check
        client.admin.command("ping")
        print("Connected to MongoDB successfully.\n")

        db_names = client.list_database_names()
        if not db_names:
            print("No databases found.")
            return

        print("Databases found:")
        for db_name in db_names:
            print(f"\n- {db_name}")
            db = client[db_name]

            collection_names = db.list_collection_names()
            if not collection_names:
                print("  No collections found.")
                continue

            print("  Collections:")
            for coll_name in collection_names:
                print(f"   - {coll_name}")
                coll = db[coll_name]

                doc_count = coll.count_documents({})
                print(f"     Document count: {doc_count}")

                sample_doc = coll.find_one()
                if sample_doc:
                    print(f"     Sample document: {sample_doc}")
                else:
                    print("     Sample document: None")

    except PyMongoError as e:
        print(f"MongoDB error: {e}")
    except Exception as e:
        print(f"Unexpected error: {e}")


if __name__ == "__main__":
    main()
```
